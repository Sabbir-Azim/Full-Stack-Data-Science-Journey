### Micro_Challenge: The Identity Swap

Intialize x=100 and y=200. Swap their values so x become 200, and y becomes 100

```python
x = 100, y = 200

x,y = y, x

print(f"After Swapping: x = {x} and y = {y}")
```

Sequences and collections

Python differentiates between ordered sequences and unordered collections (such as set and dict).

strings (str, bytes, unicode) are sequences

reversed: A reversed order of str with reversed function

```python
a = reversed('hello')
a
```

tuple: An ordered collection of n values of any type (n >= 0).

```python
a = (1, 2, 3)
b = ('a', 1, 'python', (1, 2))
b[2] = 'something else'  # immutable, not allowed
```

list: An ordered collection of n values (n >= 0)

```python
a = [1, 2, 3]
b = ['a', 1, 'python', (1, 2), [1, 2]]
b[2] = 'something else' # mutable, allowed
```

set: An unordered collection of unique values.

```python
a = {1, 2, 'a'}
```

dict: An unordered collection of unique key-value pairs;

```python
a = {1: 'one',
2: 'two'}
b = {'a': [1, 2, 3],
'b': 'a string'}
```
