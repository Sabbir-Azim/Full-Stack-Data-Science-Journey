### The infinite input pattern

```python
while True:
    pwd=input("set password(min 5 characters):")
    if len(pwd)>=5:
        print("password set")
        break      # exits the loop
    print("Too short. Try again.")
```

### Micro-Challenge: The Infinite Gurdian

```python
while True:
    password=input("set your password:")
    if password == "stop":
        break      # exits the loop
    print("Type again.")
```

### Micro-Challenge: The Accumulator Problem

```python
N = int(input("Type your number"))

sum = 0

for i in range(1, N+1):
    sum += i

print(f"The sum of all numbers from 1 to {N} is {sum}")
```

### Micro-Challenge: The Efficient Skipper

```python
for i in range(1, 11):
    if i==5:
        continue
    else:
        print(i, end="\n")

```

### Micro-Challenge: The String Walker

```python
word = "DATA"

for char in word:
    print(char, end="\n")
```
