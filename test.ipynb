```python
word = "banana"

freq = {}

for letter in word:
    if letter in freq:
        freq[letter] += 1
    else:
        freq[letter] = 1

print(freq)
```

```python
defaults = {"theme": "light", 
            "audio": "on"}

user_pref = {"theme": "dark"}

print(defaults | user_pref)
```

```python
defaults = {"theme": "light", 
            "audio": "on"}

user_pref = {"theme": "dark"}

defaults.update(user_pref)

print(defaults)
```
